```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
# Create a random erdos-renyi graph with 100 nodes and a edge probability of 0.3
ER1 = nx.erdos_renyi_graph(100, 0.3) 
ER1_DC = nx.degree_centrality(ER1) 
plt.hist(list(ER1_DC.values()))
plt.title("Degree Centrality distribution for Erdős-Rényi graph with edge probability of 0.3 - ER1")
plt.xlabel("Degree Centrality")
plt.ylabel("Occurrences")
plt.show() 
```

```python
# Change the edge probability of the erdos-reyni graph to 0.6 and plot the degree centrality distribution again.
ER2 = nx.erdos_renyi_graph(100,0.6) 
ER2_DC = nx.degree_centrality(ER2) 
plt.hist(list(ER2_DC.values())) 
plt.title("Degree Centrality distribution for Erdős-Rényi graph with edge probability of 0.6 - ER2")
plt.xlabel("Degree Centrality")
plt.ylabel("Occurrences")
plt.show() 
```

## Why does the degree centrality distribution change? Infer the changes in the plots.

The *degree centrality* ([reference](https://networkx.org/documentation/stable/reference/algorithms/generated/networkx.algorithms.centrality.degree_centrality.html#networkx.algorithms.centrality.degree_centrality)) for a node is the fraction of nodes connected to it. Therefore, higher the number of edges associated with a node, the higher degree centrality. To increase the degree centrality for a given node, we would either need to increase the number of edges in the graph while keeping the same number of nodes, or increase the number of nodes while keeping the same number of edges.

We can observe that for **ER1**, the range of degree centrality is from ~0.20 to ~0.45 and peaks at ~0.3 (the set egde probablity). Similarly, for **ER2**, the range of degree centrality is from ~0.48 to ~0.73 and peaks at ~0.6 (the set edge probablity). The peaks on both these histograms represents the most common degree centrality from nodes of the respective graph.

Based on the above observations, the degree centrality values of **ER2** are higher than that of **ER1**. Since both these graph have the same number of nodes i.e. 100, and egde probablities of 0.3 and 0.6 respectively, we can expect the number of edges in **ER2** to be greater than **ER1**. 

To verify this inference, we can compare the total number of edges in **ER1** and **ER2**.

```python
print("Total number of nodes in ER1 are: ", ER1.number_of_edges())
print("Total number of nodes in ER2 are: ", ER2.number_of_edges())
```

```python
# Create a random barabasi_albert_graph with 100 nodes and attach 3 edges to a new node in the graph (m=3). Plot the degree_centrality distribution.
BA = nx.barabasi_albert_graph(n= 100, m=3) 
BA_DC = nx.degree_centrality(BA) 
plt.hist(list(BA_DC.values())) 
plt.title("Degree Centrality distribution for Barabási–Albert graph with m=3 - BA")
plt.xlabel("Degree Centrality")
plt.ylabel("Occurrences")
plt.show() 
```

## Why does the degree centrality distribution change between these 2 random graph generators? Infer the changes in the plots.

On comparing the degree centrality distribution of these graph generators, we see that both are drastically different in nature. 

*Erdos-Renyi* graph has its distribution similar to a normal distribution (bell curve), showing that data near the mean (in our case the edge probablity) are more frequent in occurrence than data far from the mean. 

Whereas, *Barabasi-Albert* graph has its distribution similar to power law distribution (Pareto distribution) where maximum nodes take up a low degree centrality and a small fraction of nodes have large degree centrality values. 

The reason behind these differences lies in how they are generated. 
- *Erdos-Renyi graph*: A graph having a fixed probability of an egde being present or absent for any two given nodes. 
- *Barabasi-Albert graph*: A graph of nodes grown by attaching new nodes each with edges that are preferentially attached to existing nodes with high degree.

Based on these points, the observations made earlier hold true. In the degree centrality distribution of *Erdos-Renyi* graph, most nodes have a similar degree. And for *Barabasi-Albert* graph, most nodes have a lower degree and very few have higher degree values.
