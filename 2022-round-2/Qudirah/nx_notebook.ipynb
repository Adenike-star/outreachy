```python
import networkx as nx
import matplotlib.pyplot as plt
```

# Erdos-Renyi Graph

Erdos-Renyi model is used for generating random graph. The n argument specifies the number of nodes and the p argument is the probability to connect any pair of nodes(edge probability). Every edge has equal probability p of existing and the graph model has a
constant and independent probability of two nodes being connected.

```python
def ERG(n,p): #where n is number of nodes and p is the edge probability
    ERG=nx.erdos_renyi_graph(n,p) #calling the erdos_renyi_function
    values=[]
    for i in (nx.degree_centrality(ERG)).values():
        values.append(i) #adding the degree of each nodes to a list
    #drawing the network graph
    nx.draw(ERG, 
            pos =nx.kamada_kawai_layout(ERG),
            with_labels=True,
            node_color = values,
            edge_color = 'k')
    plt.title("Erdos-Renyi Graph")
    plt.show()
    plt.hist(values) #plotting the histogram of degrees
    plt.title("Histogram of Erdos-Renyi Graph with "+ str(n) + "nodes and "+ str(p) +" probability")
    plt.show()
```

```python
ERG(100,0.3) #calling the erdos-renyi graph
```

```python
ERG(100,0.6)
```

##### The diagram of the network shows that with an increase in probability, leads to a more denser graph. This is because the chances are higher for the nodes to connect. The histogram proves this statement as even though the number of nodes are the same, there seem to be more points in the second graph as a result of more nodes being connected (due to a higher probability).

# Barabasi-Albert Graph

Barabasi-Albert Graph is known to also be an algorithm that generates random graph. The graph exhibits preferential attachment as the nodes prefer to link to highly connected nodes. Because nodes are added to the network overtime, it is known for it's scale free property. Scale free network has a degree distribution which follows a power law. This means the first set of nodes can easily become the highly connected and they tend to continue to acquire more connections.

```python
def BAG(n,m): #where n is number of nodes and m is the number of edges to attach from a new node to existing nodes
    BAG=nx.barabasi_albert_graph(n,m) #calling the barabasi albert graph
    values=[] 
    for i in (nx.degree_centrality(BAG)).values():
        values.append(i) #adding the degree of each nodes to a list
    #drawing the network model    
    nx.draw(BAG, 
            with_labels=True,
            node_color = values,
            pos =nx.kamada_kawai_layout(BAG),
            edge_color = 'k')
    plt.title("Barabassi-Albert Graph")
    plt.show()
    plt.hist(values) #plotting the histogram of the degrees
    plt.title("Histogram of Barabassi-Albert graph with "+str(n)+" nodes and "+str(m)+" edges")
    plt.show()
```

```python
BAG(100,3)
```

##### The network graph shows that early nodes have more edges linked to them and later nodes (the ones with high numbers) have few edges attached to them. The histogram shows that the first set of nodes were preferred and seem to be highly linked to other set of nodes. Hence, the distribution is skewed to the right unlike the Erdos-Renyi graph which is more distributed to both sides.
