## The Erdos-Renyi Graph and Barabasi-Albert Graph
This notebook creates an erdos-renyi graph and a barabasi-albert graph
and plots the degree of centrality of the graphs.

Let us begin with the Erdos-Renyi graph.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Step 1: Create an erdos-renyi graph with 100 nodes and edge probability of 0.3
ER = nx.erdos_renyi_graph(n=100, p=0.3)
```

```python
# Step 2: obtain the degree centrality of the erdos-renyi graph and 
# plot a histogram of the degree_cenrality
deg_cen = nx.degree_centrality(ER)

plt.hist(deg_cen.values()) # plot values on a histogram
plt.title('Histogram of Degree_Centrality for Erdos-Renyi graph')
plt.xlabel('Degree_Centrality')
plt.ylabel('Number of occurencies')
plt.show()
```

```python
# Step 3: Change the edge probability of the erdos-reyni graph to 0.6 and plot the
#  degree centrality distribution again.
ER2 = nx.erdos_renyi_graph(n=100, p=0.6)

deg_cen2 = nx.degree_centrality(ER2)
plt.hist(deg_cen2.values())
plt.title('Histogram of Degree_Centrality for Erdos-Renyi graph')
plt.xlabel('Degree_Centrality')
plt.ylabel('Number of occurencies')
plt.show()
```

## Inference from the two erdos-renyi graphs with different edge probabilities
A brief introduction of edge probability and degree centrality

**Edge Probability** is the likelihood that any two random nodes in a graph will be connected through an adge.

**Degree Centrality** of a node, *v* is the fraction of total edges connected to *v*.
It can be infered from the histogram of the erdos-renyi graph with probability of 0.3 that mmost of the degree centrality was clustered around 0.3.
Similarly, when the edge probability was changed from 0.3 to 0.6, the mean of the degree centrality shifted towards 0.6.

This behaviour is caused by the edge probability. The edge probility is the likelihood that any two random nodes in a graph are going to be connected via an edge. If the probability is low, it means that nodes are going to have less and less likelihood to be connected to others. Less connections between nodes means less degree centrality.

When the edge probability is increased, the likelihood of creating edges between nodes also increases. This means more connections between nodes and hence higher degree centrality. This shifts the mean of the histogram towards 0.6.

```python
# Step 4: Create a random barabasi_albert_graph with 100 nodes and attach 3 edges 
# to a new node in the graph (m=3)
BA = nx.barabasi_albert_graph(n=100, m=3)
```

```python
# Step 5: Plot the degree_centrality distribution (a histogram of the degree histogram).
deg_cen3 = nx.degree_centrality(BA)

plt.hist(deg_cen3.values()) # Plot the histogram
plt.title("Degree Centrality for Barabasi_albert graph")
plt.ylabel('Number of Occurencies')
plt.xlabel('Degree Centrality')
plt.show()
```

## Barabasi Albert Graph
In an erdos-renyi graph, nodes are first created and edges created between any two random nodes through a given probabality. The number of edges is determined by the edge probability.

In a barabasi albert graph, each new node is created with *m* (usually specified), edges which are connected to the existing nodes. The edges are connected preferentially to existing nodes with high degree. This means that the first few nodes to get edges will increase in degree while nodes without initial edges will remain stagnant. The barabasi albert graph creates a few hightly connected nodes with the majority of nodes poorly connected.
This behaviour is evident in the histogram as the number of nodes with high degree centrality is low and vice versa.

In an erdos-renyi graph, the degree centrality is mostly clustered around the given edge probability, which is explained by the randomness of the edge creation.
