```python
# load required libraries
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns
```

## Basis

*Degree Centrality* :- The degree_centrality of a node say A is the measure of the number of edges connected to A scaled to a value between 0 and 1, where the scaling factor is the maximum possible number of edges a node can have in a network. 

ie.
$$
D(A) =  \frac{n}{N-1}
$$
 where 
        
        D(A) = degree of centrality of node A

        n = number of edges connected(drawn) to(from) A

        N = total number of nodes in the network

*Erdos–Rényi Network* :- an Erdos–Rényi Network of nodes N is a network generated by laying down N number of nodes and adding edges between them with independent probability p for each node pair.

*Barabasi_Albert Network* :- A barabasi_albert network of nodes N canbe grown by first laying down few number of connected nodes then adding nodes incremently till we reach the required N nodes. Each new node being added is connected to m existing nodes with a probability that is directly proportional to the number of edges that the existing nodes already have. ie. nodes with high degree attain even higer degree as new nodes are being added.

## Tasks



1.   Create a random erdos-renyi graph with 100 nodes and a edge probability of 0.3


```python
G1 = nx.erdos_renyi_graph(100, 0.3) # creates erdos_renyi_graph with 100 nodes and edge probility of 0.3
plt.figure(figsize=(5,5))
nx.draw(G1, with_labels=True, font_weight='bold') # visualize graph
```



2.   Plot the degree_centrality distribution (a histogram of the degree centrality values).



```python
H1 = nx.degree_centrality(G1) # return a dict containing each node and it's degree centrality
plt.title('Histogram Showing the Distribution of Degree_Centrality in a random erdos_renyi graph with n=100 nd p = 0.3')
sns.histplot(x = H1.values()) 
plt.show()
```

3. Change the edge probability of the erdos-reyni graph to 0.6 and plot the degree centrality distribution again.

```python
G2 = nx.erdos_renyi_graph(100, 0.6) 
plt.figure(figsize=(5,5))
nx.draw(G2, with_labels=True, font_weight='bold') 
```

```python
H2 = nx.degree_centrality(G2)
plt.title('Histogram Showing the Distribution of Degree_Centrality in a random erdos_renyi graph with n=100 nd p = 0.6')
sns.histplot(x = H2.values())
plt.show()
```

4. Infer the changes in the plots, why does the degree centrality distribution change. Create a markdown cell in the jupyter notebook and write what you can infer from the plots.

```python
## to compare results of two histograms
## its good practice to place them side by side(place graphs vertially to compare horizontal change and horizontally otherwise))
## use same bin width and bin range

plt.figure(figsize=(7,10))
ax1 = plt.subplot(211)
ax2 = plt.subplot(212)
sns.histplot(x = H1.values(), ax = ax1, binrange = (0,1), binwidth = 0.05)
sns.histplot(x = H2.values(), ax = ax2, binrange = (0,1), binwidth = 0.05)
ax1.set_title('Histogram Showing the Distribution of Degree_Centrality in a random erdos_renyi graph with n=100 and p = 0.3')
ax2.set_title('Histogram Showing the Distribution of Degree_Centrality in a random erdos_renyi graph with n=100 and p = 0.6')
plt.show()
```

**Inference**: 
We can see a significant change in density center distribution range between the two plots. This is because as the edge probability(p) of erdos_renyi_graph increases, the higher the number of edges created among nodes, which in turn translates to higher degree centrality.

*Note*: an erdos_renyi network with p = 1 will each node, connected to every other node.

*Other things to note*:


*   Both histograms are symmetric. This is because there is equal probability of adding edges between nodes
*   The mean of the distribution is approximately equal to the edge probability of the erdos_renyi_graph.
ie $$  u(mean) \approx p(prob-of-edge-creation) $$



5. Create a random barabasi_albert_graph with 100 nodes and attach 3 edges to a new node in the graph (m=3).

```python
G3 = nx.barabasi_albert_graph(100,3)
plt.figure(figsize=(5,5))
nx.draw(G3, with_labels=True, font_weight='bold') # visualize graph
```

6. Plot the degree_centrality distribution (a histogram of the degree histogram).

```python
H3 = nx.degree_centrality(G3) 
plt.title('Histogram Showing the Distribution of Degree_Centrality in a random barabasi_albert graph with n=100 nd m = 3')
sns.histplot(x = H3.values())
plt.show()
```

7. Infer the changes in the plots, why does the degree centrality distribution change between these 2 random graph generators Create a markdown cell in the jupyter notebook and write what you can infer from the plots.

```python
## again to compare results of two histograms, its good practice to place then side by side, use same bin width and bin range

plt.figure(figsize=(7,10))
ax1 = plt.subplot(211)
ax2 = plt.subplot(212)
sns.histplot(x = H2.values(), ax = ax1, binrange = (0,1), binwidth = 0.05) # you can use any of the two erdos_renyi graphs above
sns.histplot(x = H3.values(), ax = ax2, binrange = (0,1), binwidth = 0.05)
ax1.set_title('Histogram Showing the Distribution of Degree_Centrality in a random erdos_renyi graph')
ax2.set_title('Histogram Showing the Distribution of Degree_Centrality in a random barabasi_albert_graph')
plt.show()
```

**Inference**


There is an abrupt change in degree centrality distribution between the two plots above. 
Erdos_renyi graph is symmetric while the barabasi albert graph is right-skewed. This is mainly due to the algorithmic difference behind network generation process of two graphs. 

For the barabasi albert graph, when we add nodes we connect it to the existing nodes with already high degree therefore the initial few nodes with high degree tend to get much higher degree( forms outliers in the degree centrality histogram) as new nodes are added leaving majority of the nodes with fewer connections(lower degree_centrality, near zero).

For the erdos_renyi graph, each pair of nodes has equal probability of being interconnected, and as a result each node has near equal degree with other nodes.
