```python
# Importing libraries
import networkx as nx
import matplotlib.pyplot as plt
```

# Erdos-Renyi graph

```python
#Creating a random erdos-renyi graph with 100 nodes and a edge probability of 0.3
erdos_graph = nx.erdos_renyi_graph(100, 0.3, seed=None, directed=False)
# drawing the graph
nx.draw(erdos_graph,with_labels=True);
# creating a title
plt.title("A random erdos-renyi graph with 100 nodes and a edge probability of 0.3");
```

# Compute the degree of centrality as histogram

```python
# using the degree_centrality method to return the dictionary keyed by node with bipartite degree centrality as the value
degree_cent=nx.degree_centrality(erdos_graph)
## graph title
plt.title('Degree Centrality Distribution of Erdos Renyi Graph with 100 nodes and 0.3 edge probability')
plt.xlabel("Degree of (v)")
plt.ylabel("Frequency")
#ploting the histogram
plt.hist(list(degree_cent.values()));
plt.show();
```

# Changing the edge probability of the erdos-reyni graph to 0.6 

```python
#Creating a random erdos-renyi graph with 100 nodes and a edge probability of 0.6
erdos_graph = nx.erdos_renyi_graph(100, 0.6, seed=None, directed=False)
# drawing the graph
nx.draw(erdos_graph,with_labels=True);
# creating a title
plt.title("A random erdos-renyi graph with 100 nodes and a edge probability of 0.6");
```

```python
# using the degree_centrality method to return the dictionary keyed by node with bipartite degree centrality as the value
degree_cent=nx.degree_centrality(erdos_graph)
## graph title
plt.title('Degree Centrality Distribution of Erdos Renyi Graph with 100 nodes and 0.3 edge probability')
plt.xlabel("Degree of (v)")
plt.ylabel("Frequency")
#ploting the histogram
plt.hist(list(degree_cent.values()));
plt.show();
```

#### why does the degree centrality distribution change

#### The degree centrality of a node is the number of edges it has and according to Jennifer Golbeck the higher the degree of centerality the more the edge is at the center

* from the plots above we see that our first graph with 0.3 edge probablity has it centrality lower than that which has it probability around 0.6. 
* for the second graph the mean is more focused to a particular central point than that which has it proballility at 0.3

### barabasi_albert_graph

### barabasi_albert_graph(n, m)
#### n = number of nodes
#### m = number of edge which connect each nodes




```python
# ploting a b-a-g graph 
bag = nx.barabasi_albert_graph(100,3)
# title of graph
plt.title('Barabasi Albert Graph with 100 nodes and 3 edges')
nx.draw(bag,with_labels=True)
plt.show()
```

```python
# using the degree_centrality method to create a bar chart for the bag grah
degree_cent=nx.degree_centrality(bag)
## graph title
plt.title('Histogram of barabasi_albert_graph with 100 nodes and 3 edge connecting the nodes')
plt.xlabel("Degree of (v)")
plt.ylabel("Frequency")
#ploting the histogram
plt.hist(list(degree_cent.values()));
plt.show();
```

### Infer the changes in the plots, why does the degree centrality distribution change between these 2 random graph generators

The Erdos-Renyi graph were connected with a probability edge which made the nodes have more connections and clustered to each other. However, in the barabasi_albert_graph connection, each node has just 3 connection to any other nodes within the graph which lead to the degree of centrality being higher and concentrated in a particular point, unlike the Erdos-Renyi graph, because the connection is built on a small set of nodes which after the connection will have a higher degree of centrality which is visible on the histogram plot with more of the starting degree having higher concentrated centrality than the remaining node point node.
