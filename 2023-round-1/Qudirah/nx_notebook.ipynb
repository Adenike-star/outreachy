```python
import networkx as nx
import matplotlib.pyplot as plt
```

# Erdos-Renyi Graph

Erdos-Renyi model is used for generating random graph. It is also known as a binomial graph. In an Erdos-Renyi graph $G_{n,p}$, n specifies the number of nodes and the p is the probability to connect any pair of nodes(edge probability). Every edge has equal probability p of existing and the graph model has a constant and independent probability of two nodes being connected. The algorithm's big O notation is $O(n^{2})$. 

```python
#plotting an Erdos-Renyi Graph with 100 nodes and 0.3 probability
n = 100
p = 0.3
seed = 4 #for reproducibility
ERG = nx.erdos_renyi_graph(n, p, seed) #calling the erdos_renyi_function
nx.draw(ERG, pos=nx.spring_layout(ERG), with_labels=True, node_color = "blue") 
```

Observing the degree of centrality of the graph i.e the distribution of edges in the graph. Nodes with higher degree of centrality tend to have more edges. 

```python
values=[]
for i in (nx.degree_centrality(ERG)).values():
    values.append(i) #adding the degree values to a list
plt.hist(values) #plotting the histogram of degrees
plt.xlabel("Degree centrality")
plt.title("Degree centrality distribution of an Erdos-Renyi-graph of 100 nodes with probabilty 0.3")
plt.show()
```

Let's plot a graph with 100 nodes but 0.6 probability to see a difference. If any. 

```python
n = 100 #number of nodes
p = 0.6 #probability
seed = 4
ERG = nx.erdos_renyi_graph(n, p, seed) #calling the erdos_renyi_function
nx.draw(ERG, pos=nx.spring_layout(ERG), with_labels=True, node_color = "blue")
```

```python
values=[]
for i in (nx.degree_centrality(ERG)).values():
    values.append(i) 
plt.hist(values) #plotting the histogram of degrees
plt.xlabel("Degree centrality")
plt.title("Degree centrality distribution of an Erdos-Renyi-graph of 100 nodes with probabilty 0.6")
plt.show()
```

Even though the graph has the same number of nodes, the one with probability 0.6 is denser than the one with probability 0.3. Let's visuaize the degree of centrality side by side to make proper comparison

```python
ERG = nx.erdos_renyi_graph(100, 0.3, seed)
ERG_ = nx.erdos_renyi_graph(100, 0.6, seed)

x = [i for i in (nx.degree_centrality(ERG)).values()]
y = [i for i in (nx.degree_centrality(ERG_)).values()]

plt.hist(x, alpha=0.5, label='x')
plt.hist(y, alpha=0.5, label='y')
plt.legend(loc='upper right')
plt.show()
```

It is evident that the $G_{100,0.3}$ centers around 0.3 while $G_{100,0.6}$ centers around 0.6. The graph centrality seems to be higher for most nodes which means an increase in probability, leads to increase in more edges being connected. As a result, a more denser graph that is centered around it's probability.

# Barabasi-Albert Graph

Barabasi-Albert Graph is known to also be an algorithm that generates random graph. The graph exhibits preferential attachment as the nodes prefer to link to highly connected nodes. Because nodes are added to the network overtime, it is known for it's scale free property. Scale free network has a degree distribution which follows a power law of the form $p(k)$ ~ $k ^{-3}$. This means the first set of nodes can easily become the highly connected and they tend to continue to acquire more connections.

```python
"""plot barabasi albert graph where n is number of nodes and m is the number of edges to attach from a new node to existing
   nodes"""
BAG = nx.barabasi_albert_graph(100,3)    
nx.draw(BAG, with_labels=True,node_color = values, pos =nx.kamada_kawai_layout(BAG), edge_color = 'k')
plt.title("Barabassi-Albert Graph")
plt.show()
```

Again, let's look at the distribution of edges (degree of centrality) in the graph

```python
values=[i for i in (nx.degree_centrality(BAG)).values()] 
plt.hist(values) #plotting the histogram of degrees
plt.xlabel("Degree centrality")
plt.title("Degree centrality distribution of a Barabasi-Albert Graph of 100 nodes and 3 edges")
plt.show()
```

The histogram shows that the first set of nodes were preferred and seem to be highly linked to other set of nodes. This makes the distribution skewed to the right. This emphasized that it is of the form power law degree distribution which is found in many real world networks such as world wide web, social networks such as twitter, transportation networks, biological networks etc. 

## Conclusion

The degree distributions of Erdos-Renyi graphs and Barabasi-Albert graphs differ due to the distinct algorithms used to construct them. Erdos-Renyi graphs assign equal probability to each edge being added or not to the graph, whereas Barabasi-Albert graphs attach edges to nodes with high degree. The main contrast between these approaches is that in Erdos-Renyi graphs, the probability of adding an edge between any two nodes is independent of all other edges, while in Barabasi-Albert graphs, the likelihood of linking two nodes depends on the degree of one node.

As a result of these fundamental differences in construction, Erdos-Renyi graphs tend to have a Binomial distribution, while Barabasi-Albert graphs display a power-law degree distribution. This means that Erdos-Renyi graphs typically have many nodes with similar degrees, whereas Barabasi-Albert graphs have a few highly connected nodes and many less connected nodes. Furthermore, while Erdos-Renyi graphs are relatively simple to generate, Barabasi-Albert graphs require more complex algorithms that take into account the network's growth over time.
