### *_Erdos-renyi Graph_*

In the G(n, p) random graph model, a graph is constructed by connecting nodes randomly. Each edge is included in the graph with probability p independent from every other edge. The higher the probability _p_ the more connected the graph will become.
  
To understand it in simpler terms, imagine a group of 10 people. The probability of 2 people being friends is p. Similarly, the erdos-renyi graph of such a situation would require 2 parameters:  
    - _n: Number of nodes (friends)_  
    - _p: Probability of 2 nodes being connected_  

Let's create a Erdos-renyi graph to understand its property in a better manner.

```python
n = 100 # number of nodes
p = 0.3 # probability of connection
seed = 1021 # to set its random state
G1 = nx.erdos_renyi_graph(n, p, seed)
```

Now, let's plot this graph to get a better visual of how it looks.

```python
nx.draw(G1, with_labels=True, node_color="yellow")
```

As we can see, due to the high degree of nodes, we have a very dense graph.

### _Degree Centrality_

Degree centrality is a measure of the importance or centrality of a node (or vertex) in a network. It is based on the idea that a node's degree, or the number of edges it has, reflects its potential for interaction or communication with other nodes in the network. Degree centrality is given by:  
&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp; $ degreeCentrality(i) = \frac{degree(i)}{n-1} $

For a erdos-renyi graph, we usually observe a _binomial distribution_ in its degree centrality curve. Let's plot it to check.

```python
degreeCentralities_G1 = nx.degree_centrality(G1)
sns.histplot(degreeCentralities_G1, shrink=0.9, color="green")
plt.xlabel("Degree Centrality")
plt.title("Degree Centrality Distribution for Erdos-Renyi Graph with 100 node and p=0.3")
plt.show()
```

Therefore, the degree centrality distribution does resemble a binomial distribution graph.

### _Barbasi-Albert Graph_

The Barabasi-Albert (BA) graph is a type of scale-free network model that was introduced by Albert-László Barabási and Réka Albert in 1999. The model is based on the idea of preferential attachment, which means that new nodes in the network are more likely to attach to nodes that already have a high degree (i.e., a large number of connections).

The process of generating a Barabasi-Albert graph starts with a small number of nodes, say $n$, that are fully connected. At each time step, a $m$ new nodes are added to the network, and they are connected to $n$ existing nodes with probability proportional to their degree. This means that nodes with a high degree have a higher probability of receiving new connections than nodes with a low degree. As the network grows, the degree distribution follows a power law, with a small number of nodes having a very high degree and most nodes having a low degree.

Let's create a Barbasi-Albert Graph with 100 nodes and $m=3$

```python
n = 100 # number of nodes
m = 3 # number of edges attached
seed = 1023 # to define the random state
B = nx.barabasi_albert_graph(n, m, seed)
```

Let's plot the graph to visualize it more accurately.

```python
nx.draw(B, with_labels=True, node_color="cyan")
```

Now let's have a look at its degree centrality distribution.

```python
degreeCentralities_B = nx.degree_centrality(B)
sns.histplot(degreeCentralities_B, shrink=0.9, color="purple")
plt.xlabel("Degree Centrality")
plt.title("Degree Centrality Distribution for Barbasi-Albert Graph with 100 nodes and m=3")
plt.show()
```

As we can see, the degree centrality distribution of a Barbasi-Albert Graph resembles a power distribution curve.

## _Final Conclusion_

- *Barbasi-Albert Graph and Erdos-Renyi Graph have different degree centrality distributions*. This is because the basic characteristics and the way in which they are built are vastly different from each other. While one depends on probability for making connections, the other depends on existing degree.
- Barbasi-Albert Graph gives us a power distribution curve while Erdos-Renyi Graph gives us a binomial distribution curve for their degree centralities.
- This means that degree centrality is an important measure to judge the topology of a graph and can be used to infer important information from it.
