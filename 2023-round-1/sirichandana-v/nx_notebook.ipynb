```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
def erdos_renyi_plt(n, edge_probability):
    G=nx.erdos_renyi_graph(n, edge_probability)
    d=nx.degree_centrality(G)
    plt.hist(d.values())
    plt.title("Degree Centrality Histogram for Erdos Renyi graph with probability {}".format(edge_probability))
    plt.xlabel("Degree Centrality")
    plt.ylabel("Counts")
    plt.show()
```

## Plotting Histograms of Degree centrality for Erdos Renyi Graphs

```python
erdos_renyi_plt(100, 0.3)
erdos_renyi_plt(100, 0.6)
```

### Edge Probability: 
It is the probability of an edge existing between any two vertices in a graph.

### Degree Centrality:  
It is the percentage of the network that the particular node is connected to.


So, when edge probability increases, the no.of edges in the graph also increase. Thereby increasing the percentage of network connected by each individual node. i.e, __if the edge probability increases, degree centrality also increases.__

This can be observed by the change in the mode of two histograms.The mode of first histogram with edge-probability 0.3 is around 0.3 while the mode of second histogram with edge-probability 0.6 is around 0.6.


```python
def barabasi_albert_plt(n, m):
    G=nx.barabasi_albert_graph(n, m)
    d=nx.degree_centrality(G)
    plt.hist(d.values())
    plt.title("Degree Centrality Histogram for Barabasi Albert graph")
    plt.show()
```

## Plotting Histogram of Degree centrality for Barabasi Albert Graph

```python
barabasi_albert_plt(100, 3)
```

In an __Erdos renyi graph__,the given no.of nodes are created first and then based on the edge probability, edges are added.

Whereas, in a __Barabasi albert graph__, nodes are grown by attaching new nodes each with edges that are preferentially attached to existing nodes with high degree, which means that more nodes get connected to the already existing node with higher degree. This makes the few initial nodes to have higher degree centrality value and the majority to have lower degree centrality value. This is why the histogram is skewed to the right.

__Hence, It can be inferred that Erdos renyi graph has more nodes with high degree centrality while a Barabasi albert graph has lesser nodes with high degree centrality.__
