# **Erdős–Rényi Graph and Barabási–Albert Graph**


This is a notebook that creates a graph of the [Erdős–Rényi Model](http://www.cs.cornell.edu/courses/cs485/2006sp/lecture%20notes/lecture1.pdf) and the [Barabási–Albert Model](https://en.wikipedia.org/wiki/Barabási–Albert_model) with the help of [NetworkX](https://networkx.org), written by Naman Priyadarshi.

Importing the Libraries required

```python
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns
```

## **Erdős–Rényi Graph**


$G_{n ,  p}$ is a random graph with *n* vertices where each possible edge has probability *p* of existing

The number of edges in a $G_{n ,  p}$ graph is a random variable with expected value $\frac{n(n-1)}{2} * p $

**Degree of Centrality**

Centrality measures are among the most widely used indices based on network data. They generally reflect a unit's prominence; degree centrality is defined as the number of links incident upon a node (i.e., the number of ties that a node has)

### **Creating a Erdős–Rényi Graph with $n$ = 100 and $p$ = 0.3**

```python
#number of nodes
n = 100

#probability of each edge
p = 0.3

ERG_1 = nx.erdos_renyi_graph(n, p)

#degree of centrality
degree_1 = nx.degree_centrality(ERG_1)
```

Calculating number of edges 'e':

```python
e = (n*(n-1))*p/2

print(e)
print(ERG_1.number_of_edges())
```

Drawing the Erdős–Rényi Graph

```python
nx.draw(ERG_1, node_color='pink', node_size=200, with_labels=True)
```

The graph is very condensed as the number of nodes and edges are significant.

**Plotting degree of certrality on a histogram**

```python
sns.histplot(degree_1, color = "pink",binwidth=0.025, alpha = 1)
plt.xlabel("Degree centrality")
plt.title("Degree centrality distribution of an Erdős–Rényi Graph of 100 nodes with p = 0.3")
plt.show()
```

### Creating another Erdős–Rényi Graph with $n$ = 100 and $p$ = 0.6

```python
#Total number of nodes
n = 100

#probability of each edge
p = 0.6

ERG_2 = nx.erdos_renyi_graph(n, p)

#degree of centrality
degree_2 = nx.degree_centrality(ERG_2)
```

Calculating number of edges 'e':

```python
e = (n*(n-1))*p/2

print(e)
print(ERG_2.number_of_edges())
```

Drawing the Erdős–Rényi Graph

```python
nx.draw(ERG_2, node_color='tan', node_size=200, with_labels=True)
```

**Plotting degree of certrality on a histogram**

```python
sns.histplot(degree_2, color = "tan",binwidth=0.025, alpha = 1)
plt.xlabel("Degree centrality")
plt.title("Degree centrality distribution of an Erdős–Rényi Graph of 100 nodes with p = 0.6")
plt.show()
```

### Inference over the distributions

In conclusion, we could see that the degree-centrality distributions of the Erdős–Rényi Graph are binomial distributions. 

The graph itself shows that graph centrality increases for the majority of nodes when $p$ is increased. This is because when $p$ rises, the graph will have more edges, making it denser and more likely to be linked.

## **Barabási–Albert Graph** 

A graph of $n$ nodes is grown by attaching new nodes each with $m$ edges that are preferentially attached to existing nodes with high degree.

### **Creating a Barabási–Albert Graph with $n$ = 100 and $m$ = 3**

```python
#number of nodes 
n = 100

#number of edges attached
m = 3 

BAG = nx.barabasi_albert_graph(n, m)

#degree of centraility
degree_3 = nx.degree_centrality(BAG)
```

Drawing the Barabási–Albert Graph

```python
nx.draw(BAG, node_color='violet', node_size=200, with_labels=True)
```

**Plotting degree of certrality on a histogram**

```python
sns.histplot(degree_3, color = "violet",binwidth=0.025, alpha = 1)
plt.xlabel("Degree centrality")
plt.title("Degree centrality distribution of an Barabási–Albert Graph of 100 nodes with p=0.6")
plt.show()
```

In the histogram, we can see that the distribution of the degree centrality looks like a power distribution $P(k) \sim k^{-\gamma}$.
