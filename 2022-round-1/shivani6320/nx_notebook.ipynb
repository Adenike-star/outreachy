```python
import networkx as nx
import matplotlib.pyplot as plt
```

### erdos_renyi_graph :

```python
#creating the graph with 100 nodes and 0.3 edge probability
er1 = nx.erdos_renyi_graph(100, 0.3)
```

```python
#plotting erdos_renyi_graph
nx.draw(er1)
```

```python
#centrality degree of er1
er1_deg_cent = nx.degree_centrality(er1)
```

```python
#plotting histogram dustribution of er1 graph
plt.figure()
plt.hist(list(er1_deg_cent.values()))
plt.show()
```

- As we can see, the histogram distribution for degree of centrality for er1 graph is **Normal Distribution**
- For this graph, Mean = Median = Mode


```python
#creating the graph with 100 nodes and 0.6 edge probability
er2 = nx.erdos_renyi_graph(100, 0.6)
```

```python
#plotting erdos_renyi_graph2
nx.draw(er2)
```

```python
#centrality degree of er2
er2_deg_cent = nx.degree_centrality(er2)
```

```python
#plotting histogram dustribution of er2 graph
plt.figure()
plt.hist(list(er2_deg_cent.values()))
plt.show()
```

- The histogram distribution for degree of centrality for er2 graph is **Multimodal Distribution** as it has more than two peaks.
- For this graph, Mean = Median = Mode

- However, as "edge probability" of graph doubles(i.e from 0.3 to 0.6), the values of "degree of centrality" also doubles.
- It can also be said that, degree of centrality is directly propotional to edge probability.

### barabasi_albert_graph :

```python
#creating the graph with 100 nodes and 3 edges to attach
ba = nx.barabasi_albert_graph(100,3)
```

```python
#degree of centrality of barabasi_albert_graph
ba_deg_cent = nx.degree_centrality(ba)
```

```python
#plotting histogram distribution
plt.figure()
plt.hist(list(ba_deg_cent.values()))
plt.show()
```

- The histogram distribution for degree of centrality for barabasi albert graph is **Positively Skewed/Right-Skewed Distribution**
- For this graph, Mean > Median > Mode
