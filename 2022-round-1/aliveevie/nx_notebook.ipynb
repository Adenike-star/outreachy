```python
# import the required libraries
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Generate the graph with n=100, and p=0.3
G = nx.erdos_renyi_graph(n=100, p=0.3, seed=None, directed=False)
nx.draw(G)
```

```python
# calculate the degree of centrality
HG = nx.degree_centrality(G)
values = [] #in same order as traversing keys
keys = [] #also needed to preserve order
# use for loop to pull out the HG keys and values
for key in HG.keys():
  keys.append(key)
  values.append(HG[key])
```

```python
values[0:5]
```

```python
# plot the Histogram
plt.hist(values, bins=10)
plt.show()
```

```python
# Generate the graph with n=100, and p=0.6
G_2 = nx.erdos_renyi_graph(n=100, p=0.6, seed=None, directed=False)
nx.draw(G_2)
```

```python
HG2 = nx.degree_centrality(G_2)
values2 = []
keys2 = []
for key in HG2.keys():
  keys2.append(key)
  values2.append(HG2[key])
```

```python
values2[0:3]
```

```python
plt.hist(values2, bins=10)
plt.show()
```

It was shown that the degree of centrality increase as the probability edge increase, the values changed due change in the probabilty edge.

```python
# The barabasi albert graph
BG = nx.barabasi_albert_graph(n=100, m=3, seed=None, initial_graph=None)
nx.draw(BG)
```

```python
# compute the degree of centrality
DC = nx.degree_centrality(BG)
values3 = []
keys3 = []
for key in DC.keys():
    keys3.append(key)
    values3.append(DC[key])
```

```python
#plot the graph
plt.hist(values3, bins=10)
plt.show()
```

This shows that the barabasi albert graph has a low degree of centrality as compared to erdos renyi graph. There no strong connection between the edges and the node, as it can depict from the chart, the degree of centrality is approaching zero
