```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from matplotlib.ticker import MaxNLocator
import scipy
import seaborn as sns
```

## Create and visualize random Erdos-Renyi Graph 

```python
def ER(nodes,p):
#   initialize erdos-renyi graph
    G = nx.erdos_renyi_graph(nodes,p, directed=True)
    # function to plot degree_centrality of erdos-renyi graph G
    d = nx.degree_centrality(G) # The degree centrality for a node v is the fraction of nodes it is connected to.
    
    fig = plt.figure(figsize = (15,10))
    colors = list(d.values())
    
#     Position nodes using Kamada-Kawai path-length cost-function.
    pos = nx.kamada_kawai_layout(G)
    nx.draw(G, 
            with_labels=True,
            pos = pos,
            node_size = 500,
            node_color = colors,
            edge_color = 'k')
    plt.title("Erdos-Renyi graph")
    plt.show()
    fig = plt.figure(figsize = (15,10))
    w = 0.01
    bins = np.arange(min(list(d.values())), max(list(d.values())) + w, w)
    plt.hist(list(d.values()),bins = bins, density = True, edgecolor = 'darkblue')
    plt.title("Degree Centrality")
    plt.gca().yaxis.set_major_locator(MaxNLocator(integer=True))
    plt.show()
    
    return(d)
```

```python
d1 = ER(100, 0.3)
```

```python
d2 = ER(100, 0.6)
```

The more connections the node has the brighter it's color, therefore we see that the nodes are connected more in the center when probability is 0.3 and a little wider when probabiltiy is 0.6, i.e. more nodes are connected to each other and graph looks more dense. 

## Comparison


```python
fig = plt.figure(figsize = (10,6))
sns.kdeplot(list(d1.values()), shade = True, label = "p=0.3")
sns.kdeplot(list(d2.values()), shade = True, label = "p=0.6")
plt.gca().yaxis.set_major_locator(MaxNLocator(integer=True))
plt.legend()
```

Both figures follow normal distribution. The lower the probability (p=0.3 in our case) of graph the more thiner the graph is, i.e. more values around center. Similar with the higher probability (p=0.6) and the data points are more distributed. 

## Create and visualize random Barabasi Albert Graph 

```python
def BAR(nodes,m):
    # Initialize Barabasi Albert Random Graph
    G = nx.barabasi_albert_graph(nodes,m)
    # Function for degree_centrality
    d = nx.degree_centrality(G)
    
    fig = plt.figure(figsize = (15,10))
    colors = list(d.values()) 
#     Position nodes using Kamada-Kawai path-length cost-function.
    pos = nx.kamada_kawai_layout(G)
    nx.draw(G, 
            with_labels=True,
            pos = pos,
            node_size = 500,
            node_color = colors,
            edge_color = 'k')
    plt.title("Barabasi Albert Random Graph")
    plt.show()
    
    fig = plt.figure(figsize = (15,10)) 
    w = 0.01
    bins = np.arange(min(list(d.values())), max(list(d.values())) + w, w)
    plt.hist(list(d.values()), bins = bins, edgecolor = "lightgreen")
    plt.title("Degree Centrality")
    plt.gca().yaxis.set_major_locator(MaxNLocator(integer=True))
    plt.show()
```

```python
BAR(100,3)
```

Only few nodes are heavily connected compared to others.
