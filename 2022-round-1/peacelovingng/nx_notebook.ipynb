## An Explanatory Jupyter Notebook of Erdos-Renyi Graph and Barabasi-Albert Graph with their Degree Centrality

### Erdos-Renyi graph which is also known as a Binomial graph generates random graphs on a fixed vertex set with a fixed number of edges. The graphs are named after Hungarian mathematicians <i>Paul Erdős</i> and <i>Alfréd Rényi</i>, who first introduced them in 1959.

```python
# Import the libraries

import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline 
```

### Create a random erdos-renyi graph with 100 nodes and a edge probability of 0.3

```python
ErG_1 = nx.erdos_renyi_graph(n=100, p=0.3)
print(ErG_1)
```

### Visualize the graph using nx.draw()

```python
nx.draw(ErG_1, with_labels=True)
```

## Plot the degree_centrality distribution (a histogram of the degree centrality values)

### The degree centrality for a node v is the fraction of nodes connected to it.

```python
ErG_1_Dc = nx.degree_centrality(ErG_1).values()
```

```python
plt.hist(ErG_1_Dc, bins=100)
plt.show()
```

### Change the edge probability of the erdos-reyni graph to 0.6 and plot the degree centrality distribution again

```python
ErG_2 = nx.erdos_renyi_graph(n=100, p=0.6)
print(ErG_2)
```

### Visualize the graph using nx.draw()

```python
nx.draw(ErG_2, with_labels=True)
```

```python
ErG_2_Dc = nx.degree_centrality(ErG_2).values()
```

```python
plt.hist(ErG_2_Dc, bins=100)
plt.show()
```

### Inferences from the changes in the plots and the reason why the degree centrality distribution change

1. The second erdos-renyi graph (<b>ErG_2</b>) with the edge probability of 0.6 has twice the number of edges of the first erdos-renyi graph (<b>ErG_1</b>) with the edge probability of 0.3

2. The nx.draw() visuals of the <b>ErG_2</b> is denser than the nx-draw of <b>ErG_2</b> due to an increase in the number of edge connections.

3. There's changes in plots of the two erdos-renyi graph (though, both distribution are binomial) due to the change in the degree centrality distribution.

4. The degree centrality distribution change due to an increase in the number of edges caused by the increase in the percentage of the edge probability from 30% to 60%. <b>ErG_1</b> has its degree centrality around the 0.30 mark on the x-axis while <b>ErG_2</b> has its degree centrality close to the 0.60 mark on the x-axis of the second plot. 

5. The edge probability is directly proportional to the number edge connections generated by an erdos-renyi graph which is also directly proportional to the degree centrality distribution. 

## Create a random barabasi_albert_graph with 100 nodes and attach 3 edges to a new node in the graph (m=3)

```python
BaG = nx.barabasi_albert_graph(n=100, m=3)
print(BaG)
```

### Visualize the graph using nx.draw

```python
nx.draw(BaG, with_labels=True)
```

### Plot the degree_centrality distribution (a histogram of the degree histogram)


```python
BaG_Dc = nx.degree_centrality(BaG).values()
```

```python
plt.hist(BaG_Dc, bins=100)
plt.show()
```

## Inferences from the changes in the plots and the reason why the degree centrality distribution change between these 2 random graph generators

1. The Erdos-Renyi graphs follow a Binomial distribution pattern while the Barabasi-Albert graph has a Power Law distribution.

2. In Barabasi-Albert graphs the earlier nodes have higher degree of edge probability because nodes are added one after the other. While for Erdos-Renyi, the probability of an edge existing between two nodes is the same for any pair of nodes in the graph. 

3. In conclusion, by comparing the degree centrality distributions of these two random graph generators, we can easily see that Erdos-Renyi graph has a distribution looks like Normal distribution. In Barabasi-Albert graph, the histogram shows that there is a small number of nodes with high degree centrality, and most of other nodes have very small degree values - a few nodes take most of edges which is Power Law Distribution.
