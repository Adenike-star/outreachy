```python
import networkx as nx
import matplotlib.pyplot as plt
```

#### erdos_renyi_graph

```python
# Create a random erdos-renyi graph with 100 nodes and a edge probability of 0.3
G1 = nx.erdos_renyi_graph(100, 0.3, directed = False)
```

```python
# plot graph G1
nx.draw(G1, with_labels = True)
```

```python
# Plot the degree_centrality distribution (a histogram of the degree centrality values)
# first, compute the degree centrality of graph G1. The function returns a dictionary but we only need the values.
G1_degree_centrality_values = nx.degree_centrality(G1).values()
# second, plot a histogram of the degree centrality values
plt.figure()
plt.hist(G1_degree_centrality_values)
plt.title("Degree Centrality Values of Erdos Renyi Graph with 100 Nodes and 0.3 Edge Probability")
plt.show()
```

```python
# Change the edge probability of the erdos-reyni graph to 0.6 and plot the degree centrality distribution again.
G2 = nx.erdos_renyi_graph(100, 0.6, directed = False)
```

```python
# plot graph G2
nx.draw(G2, with_labels = True)
```

```python
# Plot the degree_centrality distribution (a histogram of the degree centrality values)
# first, compute the degree centrality of graph G2. The function returns a dictionary but we only need the values.
G2_degree_centrality_values = nx.degree_centrality(G2).values()
# second, plot a histogram of the degree centrality values
plt.figure()
plt.hist(G2_degree_centrality_values)
plt.title("Degree Centrality Values of Erdos Renyi Graph with 100 Nodes and 0.6 Edge Probability")
plt.show()
```

#### Infer the changes in the plots, why does the degree centrality distribution change. Create a markdown cell in the jupyter notebook and write what you can infer from the plots.

- Degree centrality is one of the metrics evaluating the importancy of a node, which can be calculated by the number of neighbors of a node divided by the total number of neighbors it could possibly have. The edge probability that we assigned is the probability for edge creation, which from my understanding, the larger the edge probability, the more edges are potentially to be created and the larger probability to have greater degree centrality.

- So from the graph, it can be seen that the x-axis are different, which in the first graph all the nodes' degree centrality is within 0-0.45 around the middle 0.3, whereas in the second graph all the nodes' degree centrality is within about 0.50-0.70 around the middle 0.6. This means the degree centrality has a positive relationship with edge probability, and the frequences of it is near normal distribution.

#### barabasi_albert_graph

```python
# Create a random barabasi_albert_graph with 100 nodes and attach 3 edges to a new node in the graph (m=3).
G3 = nx.barabasi_albert_graph(100, 3)
```

```python
# plot graph G3
nx.draw(G3, with_labels = True)
```

```python
# Plot the degree_centrality distribution (a histogram of the degree histogram).
# first, compute the degree centrality of graph G2. The function returns a dictionary but we only need the values.
G3_degree_centrality_values = nx.degree_centrality(G3).values()
# second, plot a histogram of the degree centrality values
plt.figure()
plt.hist(G3_degree_centrality_values)
plt.title("Degree Centrality Values of Barabasi Albert Graph with 100 Nodes and 0.3 Edge Probability")
plt.show()
```

#### Infer the changes in the plots, why does the degree centrality distribution change between these 2 random graph generators Create a markdown cell in the jupyter notebook and write what you can infer from the plots.

- It can be seen that the barabasi_albert graph is extremely right-skewed, whereas the erdos_renyi graph is more likely to normal distribution. 

- This is because when generating the graphs, we are using two completely different techniques. According to the networkX documentation, the 'p' parameter in erdos_renyi_graph() function means the probability for edge creation, namely, for each edge with this node, it has the same probability of being created or not. While the 'm' parameter inside barabasi_albert_graph() function holds the meaning of how many edges to attach from the new node to existing ndoes, namely, the preferential number of edges attached to existing nodes. That is to say, if an existing node has enough edges, the new node could possibly not be able to attach new edges to this node, which causes the graph right-skewed as less edges being added and less nodes being of importance. Therefore, the number of edges in barabasi_albert_graph seems to be lesser than those in erdos_renyi_graph, and the degree centrality is much smaller.
