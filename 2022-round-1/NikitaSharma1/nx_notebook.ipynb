```python
#importing libraries
import networkx as nx
import matplotlib.pyplot as plt
```

### Erdos renyi graph with 100 nodes and 0.3 probability

```python
ERG1= nx.erdos_renyi_graph(100,0.3)
degree_cent1= nx.degree_centrality(ERG1)
```

```python
#plotting histogram of degree centrality of erdos renyi graph

fig1, ax1 = plt.subplots(figsize =(7, 5))
ax1.hist(list(degree_cent1.values()), color='#96EEE5',edgecolor='#8C9291')
ax1.set_title("Histogram of Degree Centrality Values", fontsize=15)
ax1.set_xlabel("Degree Centrality", fontsize=13)
ax1.set_ylabel("Frequency",fontsize=13)
plt.show()
```

### Erdos renyi graph with 100 nodes and 0.6 probability

```python
ERG2= nx.erdos_renyi_graph(100,0.6)
degree_cent2= nx.degree_centrality(ERG2)
```

```python
#plotting histogram of degree centrality of erdos renyi graph

fig2, ax2 = plt.subplots(figsize =(7, 5))
ax2.hist(list(degree_cent2.values()), color='#E8EF7A',edgecolor='#8C9291')
ax2.set_title("Histogram of Degree Centrality Values", fontsize=15)
ax2.set_xlabel("Degree Centrality", fontsize=13)
ax2.set_ylabel("Frequency",fontsize=13)
plt.show()
```

#### Conclusion

<b>erdos_renyi_graph(n,p,seed=None,directed=False):</b> Erdos renyi graph is a random graph which connects each pair of nodes(edge) with probability p. The more the probabilty the more are the chances of having more edges.<br>
<b>Degree Centrality:</b> It is the number of edges connected to a node. nx.degree_centrality() returns the centrality after normalising it by dividing the degree of node with the maximun possible degree (n-1).<br><br>
Here, in ERG1, the frequency of nodes having around 0.3 degree centrality is more that means most of the nodes are connected to approx 30 other nodes in this case.<br> While in ERG2, the frequency of nodes having around 0.6 degree centrality is more that means most of the nodes are connected to approx 60 other nodes in this case.<br> This difference occured because the probability given in erdos renyi graph function in ERG1 is less than that of in ERG2.

### Barabasi Albert Graph with 100 nodes and 3 edges per new node

```python
BAG=nx.barabasi_albert_graph(100,3)
bag_degree_cent=nx.degree_centrality(BAG)
```

```python
#plotting histogram of degree centrality of Barabasi Albert Graph

fig3, ax3 = plt.subplots(figsize =(7, 5))
ax3.hist(list(bag_degree_cent.values()), color='#EF817A',edgecolor='#8C9291')
ax3.set_title("Histogram of Degree Centrality Values", fontsize=15)
ax3.set_xlabel("Degree Centrality", fontsize=13)
ax3.set_ylabel("Frequency",fontsize=13)
plt.show()
```

#### Conclusion

<b>barabasi_albert_graph(n, m, seed=None, initial_graph=None):</b> A graph of n nodes is grown by attaching new nodes each with m edges that are preferentially attached to existing nodes with high degree.<br><br>
Here, in BAG, every new node is attaching itself to 3 high degree nodes (m=3) which are mainly the first occuring nodes as you can see in the adjacency list below. The nodes 0,4,9 and 10 have more nodes attached to them than the other nodes. It actually kind of satisfies the logic of the rich getting richer and the poor getting poorer. Hence, the frequency of nodes with higher degree centrality is less as compared to those with lower degree centrality.<br><br>

The difference between Erdos Renyi Graph(ERG) and Barabasi Albert Graph(BAG) is that in ERG, each node have equal probaility of connecting while in BAG higher degree nodes have more probability.
