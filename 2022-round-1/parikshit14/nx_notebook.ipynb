```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.erdos_renyi_graph(n=100, p=0.30, directed=False)
```

```python
nx.draw(G,with_labels=True)
```

```python
G_dc = nx.degree_centrality(G).values()
```

```python
plt.figure()
plt.hist(G_dc)
```

```python
G_new = nx.erdos_renyi_graph(n=100, p=0.60, directed=False)
```

```python
nx.draw(G_new,with_labels=True)
```

```python
G_new_dc = nx.degree_centrality(G_new).values()
```

```python
plt.figure()
plt.hist(G_new_dc)
```

### Observations:
Since we are changing the probability on number of edge in a **erdos_renyi_graph** from 30% to 60% we get changes in number of edges and hence no. of edges associated with each node also changes(increases) hence **degree_centrality** also changes.

```python
B = nx.barabasi_albert_graph(n=100,m=3)
```

```python
B.add_edge(2,10)
```

```python
nx.draw(B,with_labels=True)
```

```python
B_dc = nx.degree_centrality(B).values()
```

```python
plt.hist(B_dc)
```

### observation
In **barabasi_albert_graph** 
- Nodes are added on a one by one basis
- Node added earlier get higher degree because probability is proportional to the degree of the existing node

Difference between the two graphs
- we get almost normal distribution in **erdos_renyi_graph** while we get a power law distribution in **barabasi_albert_graph**
