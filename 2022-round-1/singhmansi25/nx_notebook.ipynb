```python
# Importing libraries
import networkx as nx
import matplotlib.pyplot as plt
```

## Erdos_Renyi_Graph

### Task:  To create a random erdos-renyi graph with 100 nodes and a edge probability of 0.3

```python
# number of nodes, n
n=100
# edge probability, p
p = 0.3
# creating Erdos_Renyi_Graph using networkx library
EG = nx.erdos_renyi_graph(n,p)
# title of graph
plt.title('Erdos Renyi Graph with 100 nodes and 0.3 edge probability')
nx.draw(EG,with_labels=True)
plt.show()
```

### Task: Plot the degree_centrality distribution (a histogram of the degree centrality values).

```python
# using degree_centrality method to compute the degree centrality for nodes in a bipartite network
# it returns a dictionary keyed by node with bipartite degree centrality as the value.
deg_dict = nx.degree_centrality(EG)
# title of graph
plt.title('Degree Centrality Distribution of Erdos Renyi Graph with 100 nodes and 0.3 edge probability',fontdict={'fontsize':15})
# x and y axis labels of graph
plt.xlabel('Degree (k)')
plt.ylabel('Fraction of Nodes')
# plotting histogram with dictionary
plt.hist(list(deg_dict.values()),edgecolor='black')
plt.show()
```

### Task:  Change the edge probability of the erdos-reyni graph to 0.6 and plot the degree centrality distribution again.

```python
# number of nodes, n
n=100
# edge probability, p
p = 0.6
# creating Erdos_Renyi_Graph using networkx library
EG1 = nx.erdos_renyi_graph(n,p)
# title of graph
plt.title('Erdos Renyi Graph with 100 nodes and 0.6 edge probability')
nx.draw(EG1,with_labels=True)
plt.show()
```

### Task: Plot the degree_centrality distribution (a histogram of the degree centrality values).

```python
# using degree_centrality method to compute the degree centrality for nodes in a bipartite network
# it returns a dictionary keyed by node with bipartite degree centrality as the value.
deg_dict = nx.degree_centrality(EG1)
# title of graph
plt.title('Degree Centrality Distribution of Erdos Renyi Graph with 100 nodes and 0. edge probability',fontdict={'fontsize':15})
# x and y axis labels of graph
plt.xlabel('Degree (k)')
plt.ylabel('Fraction of Nodes')
# plotting histogram with dictionary
plt.hist(list(deg_dict.values()),edgecolor='black')
plt.show()
```

For first erdos renyi graph, EG:

n = 100, p = 0.3

The average number of edges for this is: \begin{equation*} {n \choose 2} p = 15\end{equation*}

For first erdos renyi graph, EG1:

n = 100, p = 0.6

The average number of edges for this is: \begin{equation*} {n \choose 2} p = 30\end{equation*}

Here p, i.e. edge probability is like a weighing function and as p increases from 0 to 1. 

### The graph model becomes more and more likely to include graphs with more edges and less and less likely to include graphs with fewer edges.  
In particular, the case p = 1/2 corresponds to the case where all 2^(nC2) graphs on n vertices are chosen with equal probability.

### Task: Create a random barabasi_albert_graph with 100 nodes and attach 3 edges to a new node in the graph (m=3).

```python
# number of nodes, n
n = 100
# number of edges connected to each node, m 
m = 3
# creating Barabasi_Albert_Graph using networkx library
BG1 = nx.barabasi_albert_graph(n,m)
# title of graph
plt.title('Barabasi Albert Graph with 100 nodes and 3 edges')
nx.draw(BG1,with_labels=True)
plt.show()
```

### Task: Plot the degree_centrality distribution (a histogram of the degree centrality values).

```python
# using degree_centrality method to compute the degree centrality for nodes in a bipartite network
# it returns a dictionary keyed by node with bipartite degree centrality as the value.
deg_dict = nx.degree_centrality(BG1)
# title of graph
plt.title('Degree Centrality Distribution of Barabasi Albert Graph with 100 nodes and 3 edges',fontdict={'fontsize':15})
# x and y axis labels of graph
plt.xlabel('Degree (k)')
plt.ylabel('Fraction of Nodes')
# plotting histogram with dictionary
plt.hist(list(deg_dict.values()),edgecolor='black')
plt.show()
```

### Task:
### Infer the changes in the plots, why does the degree centrality distribution change between these 2 random graph generators 
### Create a markdown cell in the jupyter notebook and write what you can infer from the plots.

The degree distribution of Erdos-Renyi Graph for small value of n is binomial distribution. 

Whereas for Barabasi Albert Graph, the distribution is scale-free.

In the Erdos-Renyi Graph, we assign n nodes, and then connect each pair with probability p. 

As edge probability increases (0 -> 1), network becomes denser.

This means that no node will have much higher degree than any other.

In the Barabasi Albert Graph, we assign n nodes, but to create them, we first start with a small set of connected nodes. 

Then we add nodes one at a time till we get n nodes. When we add a node, we connect it to a small number of existing nodes with probability proportional to the degree of the existing node. 

As a result, nodes with higher degree (the earlier ones) tend to get even higher degree.
