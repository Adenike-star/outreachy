```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
er_graph_1=nx.erdos_renyi_graph(100, 0.3)
```

```python
nx.draw(er_graph_1,with_labels=1)
```

```python
dcen_er_g1=nx.degree_centrality(er_graph_1)
print(dcen_er_g1)
```

```python
plt.hist(list(dcen_er_g1.values()))
plt.title("Degree Centrality of er_graph_1")
plt.show()
```

```python
er_graph_2=nx.erdos_renyi_graph(100, 0.6)
```

```python
nx.draw(er_graph_2,with_labels=1)
```

```python
dcen_er_g2=nx.degree_centrality(er_graph_2)
print(dcen_er_g2)
```

```python
plt.hist(list(dcen_er_g2.values()))
plt.title("Degree Centrality of er_graph_2")
plt.show()
```

The difference in the above two histograms is because erdos_renyi_graph(), gives a random graph as an output and nodes connection (edges) with the given probability. Hence, when the first erdos_renyi_graph is generated it has given a random graph with the possibility of edges at 0.3 and for the second execution, it has generated a random graph along with edge connectivity at a probability of 0.6. The degree of centrality value is gievn by is input (for differnt values, different output.)
